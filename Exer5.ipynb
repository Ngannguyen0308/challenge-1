```python
import time
class Timer(object):
    def __init__(self):
        self.times = []
        self.start()

    def start(self):
        self.start_time = time.time()

    def stop(self):
        self.times.append(time.time() - self.start_time)
        return self.times[-1]
```

```python
timer = Timer()
l = input("Enter your user id")
def tran_follow_id(l):
    res = 0
    for i in range(len(df)):
        if df.iloc[i]['user_id']== l:
            res += 1
        
    print('Time to exec mean_func %.5f sec:' % timer.stop())
    return res
print(tran_follow_id(l))
```
